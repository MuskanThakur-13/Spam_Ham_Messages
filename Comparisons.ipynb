```python
import pandas as pd
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
files=files.upload()
```

```python
BOW_Stem=pd.read_csv('Bag of words using stemming.csv')
```

```python
BOW_Stem.head()
```

```python
BOW_Stem=BOW_Stem.drop(['Unnamed: 0'],axis=1)
```

```python
BOW_Stem.head()
```

```python
#Graph plotting using Bag Of Words using Stemming
Model=BOW_Stem['Model']
Accuracy=BOW_Stem['Accuracy (test set)']
Time=BOW_Stem['Training time (sec)']
Precision=BOW_Stem['Precision']
Recall=BOW_Stem['Recall']
F1_Score=BOW_Stem['F1 score']
F2_Score=BOW_Stem['F2 score']
ROC_AOC_Score=BOW_Stem['ROC-AUC score']

X=np.arange(len(Accuracy))
X2=X
W=0.1
bar1=plt.bar(X,Accuracy,width=W)
X=X+W
bar2=plt.bar(X,Time,width=W)
X=X+W
bar3=plt.bar(X,Precision,width=W)
X=X+W
bar4=plt.bar(X,Recall,width=W)
X=X+W
bar5=plt.bar(X,F1_Score,width=W)
X=X+W
bar6=plt.bar(X,F2_Score,width=W)
X=X+W
bar7=plt.bar(X,ROC_AOC_Score,width=W)

plt.xticks(X2+(W+(W/2)),Model)
plt.title("Bag of Words using Stemming")
plt.legend( (bar1, bar2, bar3,bar4, bar5, bar6,bar7), ('Accuracy', 'Time', 'Precision','Recall', 'F1_Score', 'F2_Score','ROC_AOC_Score') )

```
